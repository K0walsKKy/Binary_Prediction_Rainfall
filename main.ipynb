```python
X = df.copy().drop(columns=['rainfall'])
y = df['rainfall']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.8, random_state=13)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state =42)
X_resampled, y_resampled = smote.fit_resample(X_train,y_train)
```

```python
scaler = StandardScaler()

s_train = scaler.fit_transform(X_resampled)
s_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix
from scipy.stats import randint
from sklearn.metrics import auc, precision_recall_curve, roc_curve, roc_auc_score, average_precision_score
```

```python
!pip install imblearn
```

```python
param_grid = {
    'penalty': ['l1', 'l2'],  # Retirez 'elasticnet' si inutile
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga'],  # 'liblinear' pour l1/l2, 'saga' pour elasticnet
    'class_weight': [None, 'balanced'],
    'max_iter': [5000],  # Augmentez le nombre d'itérations
    'tol': [1e-4]  # Tolérance pour arrêter plus tôt
}

# Création du modèle
model = LogisticRegression()

grid = GridSearchCV(model, param_grid=param_grid, cv=10, n_jobs=-1)

grid.fit(s_train, y_resampled)


print(f"Meilleurs hyperparamètres pour le LogisticR : {grid.best_params_}")
```

```python
lr = LogisticRegression(
    C = grid.best_params_['C'],
    penalty=grid.best_params_['penalty'],
    solver=grid.best_params_['solver'],
    class_weight = grid.best_params_['class_weight'],
    max_iter = grid.best_params_['max_iter'],
    tol = grid.best_params_['tol']
)

lr.fit(s_train, y_resampled)
```

```python
y_pred = lr.predict(s_test)
y_score = lr.predict_proba(s_test)[:, 1]

roc_auc = roc_auc_score(y_test, y_score)
avg_precision = average_precision_score(y_test, y_score)


print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print(f"ROC AUC Score: {roc_auc:.2f}")
print(f"Average Precision Score: {avg_precision:.2f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = {
    'n_estimators': randint(50, 300),           # 50 à 300 arbres
    'max_depth': [5, 10, 20, None],             # Contrôle la complexité
    'min_samples_split': randint(2, 20),        # Évite le surapprentissage
    'min_samples_leaf': randint(1, 10),         # Évite les feuilles trop petites
    'max_features': ['sqrt', 'log2', 0.8],      # Diversité des arbres
    'class_weight': [None, 'balanced', {0:1, 1:5}]  # Ponde la classe minoritaire
}

# Créer le modèle
rf = RandomForestClassifier(random_state=42)

# Lancer la recherche (50 itérations pour rapidité)
random_search = RandomizedSearchCV(
    rf, 
    param_distributions=param_dist,
    n_iter=50,
    cv=5,
    scoring='average_precision',  # Optimise l'AP (adapté au déséquilibre)
    n_jobs=-1,                    # Utilise tous les cœurs CPU
    random_state=42
)

random_search.fit(s_train, y_resampled)

print("Meilleurs hyperparamètres :", random_search.best_params_)
```

```python
y_pred = random_search.predict(s_test)
y_score = random_search.predict_proba(s_test)[:, 1]

roc_auc = roc_auc_score(y_test, y_score)
avg_precision = average_precision_score(y_test, y_score)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print(f"ROC AUC Score: {roc_auc:.2f}")
print(f"Average Precision Score: {avg_precision:.2f}")
```
